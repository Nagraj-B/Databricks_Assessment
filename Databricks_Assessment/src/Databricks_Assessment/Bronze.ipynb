```python
from pyspark.sql.functions import *
```

```python
dbutils.widgets.text("catalog","Databricks_QA")  
```

```python
catalog=dbutils.widgets.get("catalog")
```

```python
df=spark.read.csv(f"/Volumes/databricks_dev/bronze/customers",header=True,inferSchema=True)
```

```python
df1=df.withColumn("ingestion_date",current_timestamp()).withColumn("path",col("_metadata.file_path"))
```

```python
df1.write.mode("overwrite").saveAsTable(f"{catalog}.bronze.Customers")
```
